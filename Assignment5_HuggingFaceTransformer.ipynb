```python
from transformers import pipeline
```

```python
#Text Generation using GPT-2

# Load text generation pipeline
generator = pipeline("text-generation", model = "gpt2")
# Example prompt
prompt = "Share a wholesome story"
# Generate text
generated_text = generator(prompt, max_length=150, num_return_sequences=1)[0]['generated_text']
# Print generated text
print("Generated text:", generated_text)
```

```python
#Text to Sentiment Analysis
emotion = pipeline('sentiment-analysis', model='arpanghoshal/EmoRoBERTa')
text = input("Write a sample sentence")
emotion_labels = emotion(text)
emotion_labels
```

```python
#Summarization with Bart

# Load summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
# Example input text
big_Essay = "The rapid advancement of technology has revolutionized every aspect of human life. From communication to healthcare, education to entertainment, technology has become an integral part of our daily existence. With the advent of smartphones and the internet, information is now accessible at our fingertips, connecting people across the globe like never before. However, this digital age also brings challenges, including concerns about privacy, cybersecurity, and the digital divide. It is imperative for society to adapt responsibly to these changes, ensuring equitable access to technology while safeguarding individual rights and promoting digital literacy. Embracing technology wisely can empower us to shape a brighter future."
# Generate summary
summary = summarizer(big_Essay)
summary
```
